```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
raw_data = pd.read_csv('1.03.+Dummies.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
res = sm.OLS(y,x).fit()
res.summary()
```

```python
plt.scatter(data['SAT'],y)
yhat_yes = 0.6439 + 0.0014 * data['SAT']
yhat_no = 0.8665 + 0.0014 * data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='orange')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='red')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
yhat_yes = 0.6439 + 0.0014 * data['SAT']
yhat_no = 0.8665 + 0.0014 * data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='orange')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='red')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
yhat_yes = 0.6439 + 0.0014 * data['SAT']
yhat_no = 0.8665 + 0.0014 * data['SAT']
yhat = 0.0017 * data['SAT'] + 0.275 
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='orange')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='red')
fig = plt.plot(data['SAT'], yhat, lw=3, c='green')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```
